```python
import requests
from pprint import pprint
from requests import exceptions # make use of exceptions module to handle exceptions with requests
```

```python
requests.get('http://nonexistentttt.com')
```

```python
try:
    requests.get('http://nonexistenttt.com')
except exceptions.ConnectionError:
    print('Error: Connection Error')
```

```python
# CONNECTION TIMEOUT ERROR
```

```python
requests.get('https://github.com', timeout=0.001)
```

```python
try:
    requests.get('https://github.com', timeout=0.001)
except exceptions.ConnectTimeout:
    print('Timeout Error')
```
