```python
import requests
from pprint import pprint
```

```python
r_post = requests.post('https://en.wikipedia.org/w/index.php',
                          data = { 'search': 'Skillsoft' })
```

```python
r_post.status_code
```

```python
pprint(r_post.url)
```

```python
type(r_post)
```

```python
pprint(r_post.text)
```

```python
# open a file using open() with 'wb' as overwritting existing file and in a binary mode
```

```python
with open('skillsoft.html', 'wb') as f:
    # iterate thru content of body in the form of chunk
    for chunk in r_post.iter_content(chunk_size=10000):
        f.write(chunk)
```
