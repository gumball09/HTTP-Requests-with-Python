```python
from PIL import Image
from io import BytesIO
import requests
```

```python
resp = requests.get('https://upload.wikimedia.org/wikipedia/commons/thumb/7/7e/Melanodryas_cucullata_-_Glen_Davis.jpg/1920px-Melanodryas_cucullata_-_Glen_Davis.jpg')
```

```python
resp.status_code
```

```python
type(resp.content)
```

```python
# Option1: Render the image within Jupyter Notebook
```

```python
image = Image.open(BytesIO(resp.content)) # Convert from binary type of JPEG type
```

```python
type(image) # check the type
```

```python
image.save('male_hooded_robin.png') # save the image as PNG file
```
